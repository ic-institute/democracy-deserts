```python
import pandas as pd

from ddeserts.annotate import add_dvap_columns
from ddeserts.annotate import add_geo_columns
from ddeserts.annotate import add_has_charter_column
from ddeserts.annotate import with_columns_sorted
from ddeserts.load import load_cvap_data
from ddeserts.load import load_charter_cities
```

```python
# read in list of charter cities
charter_cities = load_charter_cities()
```

```python
# load the data
pre_filter = lambda line: ('California' in line and 'Total' in line)
filter = lambda row: (row['state'] == 'California' and 
                      row['geotype'] != 'cdp' and
                      row['lnnumber'] == 1)

df = load_cvap_data('Place', pre_filter=pre_filter, filter=filter)
```

```python
# add extra columns
add_dvap_columns(df)
add_geo_columns(df)
add_has_charter_column(df, charter_cities)
df = with_columns_sorted(df)
df
```

```python
df.to_csv('output/ca-cities.csv')
```
